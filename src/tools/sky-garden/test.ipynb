```python

def client_to_bot(client_x, client_y, client_width, client_height):
    x = round((client_x / client_width) * 100, 2)
    y = round((client_y / client_height) * 100, 2)

    return {
        "x": x,
        "y": y
    }


# =====================
# EXAMPLE
# =====================

# lấy từ Window Spy
client_x = int(input())
# client_x = 642
client_y =int(input())
# client_y =152

# kích thước vùng game
client_width = 840
client_height = 485

result = client_to_bot(
    client_x,
    client_y,
    client_width,
    client_height
)

print(result)

# output:
# {'x': 39.38, 'y': 94.44}

```
